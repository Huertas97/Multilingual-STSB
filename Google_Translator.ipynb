### Google Translator

```python
from deep_translator import GoogleTranslator
# Sentence to translate
s = "keep it up, you are awesome"
# Translators
Google_trans = GoogleTranslator(source='auto', target='es').translate(s) 


print(f"Original sentence >>> {s}")
print(f"Google Translator >>> {Google_trans}")
```

### DeepL

In order to use the DeepL translator, you need to generate an api key. Deepl offers a Pro and a free API. deep-translator supports both Pro and free APIs. Just check the examples below. Visit https://www.deepl.com/en/docs-api/ for more information on how to generate your Deepl api key

DeepL free API use conditions:

Your subscription to DeepL API Free is completely free.
You can translate up to 500,000 characters per month.
You can upgrade to DeepL API Pro at any time

```python
# Get your API key
with open("DeepL_key.txt", "r") as f: # Insert your own key
  k = f.readlines()[0].strip()
```

```python
from deep_translator import DeepL
DeepL_trans = DeepL(api_key=k, source="en", target="es", use_free_api=True).translate(s)
print(f"Original sentence >>> {s}")
print(f"DeepL Translator >>> {DeepL_trans}")
```
